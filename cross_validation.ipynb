```python
import pandas as pd
import os
os.chdir("C:\\Users\\stella\\Desktop\\deploiement\\eoc-ai-session-2-main\\data")
data = pd.read_csv('breast_cancer.csv', sep = ';')
print('data', data.shape)
```

```python
data.groupby(["pam50"]).size()
```

```python
#donnees d'expression de 50 genes
X = data.select_dtypes('number')
```

```python
X
```

```python
#Etiquettes correspondantes
y = data['pam50']
print('y', y.shape)
```

```python
from sklearn.model_selection import StratifiedKFold

random_state = 0
cross_validation = StratifiedKFold(n_splits = 3, random_state =random_state, shuffle = True)

for train_index, test_index in cross_validation.split(X, y):
    X_train = X.iloc[train_index]
    X_test = X.iloc[test_index]
    print('Train', X_train.shape, 'Test', X_test.shape)
```

***Calcul detaille***

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn import metrics

scaler = StandardScaler()
classifier = SVC(kernel = 'linear', random_state= random_state, class_weight = 'balanced')

all_score = pd.Series(dtype = float)

iteration = 0
for train_index, test_index in cross_validation.split(X, y):
    iteration += 1
    
    #train dataset
    X_train = X.iloc[train_index]
    y_train = y.iloc[train_index]
    
    #test dataset
    X_test = X.iloc[test_index]
    y_test = y.iloc[test_index]
    
    #normaliser les donnees(scaled data) centree reduite
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    #train
    classifier.fit(X_train_scaled, y_train)
    
    #prediction
    y_pred_test = classifier.predict(X_test_scaled)
    
    #Accuracy
    all_score_test= metrics.accuracy_score(y_test, y_pred_test)
    all_score.loc[iteration]= all_score_test
    
    #afficher
    print('Iteration', iteration, 'Accuracy=', '{:.8f}'.format(all_score_test))
print(f"Mean accuracy  {all_score.mean()}")    
    
    
    
```

```python
 y_pred_test
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score

pipeline = Pipeline([('scaler', scaler), ('classifier', classifier)])

all_score = cross_val_score(pipeline, X, y, cv =cross_validation)
print(all_score)
print(f"Mean accuracy  {all_score.mean()}") 
```
